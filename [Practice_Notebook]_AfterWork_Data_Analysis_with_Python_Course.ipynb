## 4. Ranking / Sorting

### Examples

```python
# Sorting in Ascending / Descending Order
# ---
# From the ebola dataset sort the Guinea records in descending order
# ---
# We filter the DataFrame 'ebola_df' to only include rows where the 'Country' column matches 'Guinea.'
# Then, we sort these filtered rows in ascending order based on the 'Value' column.
#
guinea_records = ebola_df[ebola_df.Country.isin(['Guinea'])].sort_values(by='Value', ascending=1)
guinea_records.head()
```

```python
# Ranking in Ascending Order
# ---
# Create a new column that is the rank of the value of 'Value' in descending order
# ---
# We calculate the rank of values in the 'Value' column of the 'ebola_df' DataFrame
# in descending order (highest to lowest). This assigns a rank to each value,
# with the highest value receiving a lower rank and the lowest value receiving the highest rank.
#
ebola_df['Ranking'] = ebola_df['Value'].rank(ascending=0)
ebola_df.head()
```

## 5. Aggregate Functions

### Examples

```python
# Example 1
# ---
# Shape
# ---
# How many Liberian records do we have in our ebola dataset?
# ---
# (records, columns)
#
liberia_records = ebola_df[ebola_df.Country.isin(['Liberia'])]
liberia_records.shape

# (9 records, 7 columns)
```

```python
# Example 2
# ---
# Count() Function
# ---
# How many values do we have across our variables in our ebola dataset?
# ---
# We filter the DataFrame 'ebola_df' to include only the rows
# where the 'Country' column matches 'Liberia'
#
liberia_records = ebola_df[ebola_df.Country.isin(['Liberia'])]
liberia_records.count()
```

```python
# Example 3
# ---
# Unique Occurrences
# ---
# Were all records in our ebola dataset unique?
# ---
# We check for duplicated rows in the DataFrame 'ebola_df' and then use .shape
# to determine the dimensions (number of rows and columns) of the resulting
# DataFrame containing the duplicated rows.
#
ebola_df[ebola_df.duplicated()].shape

# Yes, there were no duplicate records
```

```python
# Example 4
# ---
# Sum
# ---
# How many ebola cases in Sierra Leone were reported given the ebola dataset?
# ---
# We filter the DataFrame 'ebola_df' to include only the rows where
# the 'Country' column matches 'Sierra Leone.' Then, we select the 'Value' column
# from these filtered rows and calculate the sum of the values in that column.
# This gives us the total sum of values for 'Sierra Leone' in the 'Value' column
# of the DataFrame.
#
sierra_cases_sum = ebola_df[ebola_df.Country.isin(['Sierra Leone'])][['Value']].sum()
sierra_cases_sum
```

```python
# Example 5
# ---
# Minimum
# ---
# Which record had the least no of ebola cases?
# ---
# We compute the minimum value for each column in the DataFrame 'ebola_df'
# using the min() function.
#
least_cases = ebola_df.min()
least_cases
```

```python
# Maximum
# ---
# Which record had the highest no of ebola cases?
# ---
# We compute the maximum value for each column in the DataFrame 'ebola_df' using the max() function
#
highest_cases = ebola_df.max()
highest_cases
```

```python
# Mean
# ---
# What was the average number of ebola cases in Sierra Leone?
# ---
# We filter the DataFrame 'ebola_df' to include only the rows where the 'Country' column
# matches 'Sierra Leone.' Then, we select the 'Value' column from these filtered rows
# and calculate the mean (average) of the values in that column.
# This provides the average value for 'Sierra Leone' in the 'Value' column of the DataFrame.
#
sierra_cases_avg = ebola_df[ebola_df.Country.isin(['Sierra Leone'])][['Value']].mean()
sierra_cases_avg
```

```python
# Mode
# ---
# What was the most frequent value in the Ebola dataset?
# ---
# We calculate the mode (the most frequently occurring value) of the 'Country' column
# in the DataFrame 'ebola_df'. This will return a series with the mode(s) of the 'Country' column,
# which represents the country or countries that appear most frequently in the dataset.
#
frequent_values_df = ebola_df['Country'].mode()
frequent_values_df
```

```python
# Median
# ---
# The median value of the Ebola cases.
# ---
# We compute the median of the 'Value' column in the DataFrame 'ebola_df'.
# The median is the middle value when the values are sorted in ascending order,
# and it represents the central value in a dataset.
#
median_values_df = ebola_df['Value'].median()
median_values_df
```

```python
# Summary Statistics
# ---
# Create a descriptive statistics table of the given ebola dataset
# in preparation for further analysis.
# ---
# We generate a summary statistical description of the DataFrame 'ebola_df' using the describe() function.
# This summary includes various statistics for each numerical column in the DataFrame,
# such as count, mean, standard deviation, minimum, 25th percentile, median (50th percentile),
# 75th percentile, and maximum. This provides an overview of the central tendency and spread
# of the numerical data in the DataFrame.
#
ebola_df.describe()
```

## 6. Grouping

### Examples

```python
# Let's first load and preview the dataset that we will need for our examples
# ---
#
insurance_df = pd.read_csv('https://bit.ly/2y5CRYc')
insurance_df.head()
```

```python
# Example 1
# ---
# The groupby() function is used to split the data into groups based on some criteria.
# ---
# We will start by grouping by a single column and including an aggregate function as shown.
# ---
# NB: We will not store any our dataframe in a variable for ease of
# performing computations
# ---
# We group the DataFrame 'insurance_df' by the 'region' column and then calculate the mean for each group.
# This operation provides the mean (average) values of all numerical columns in the DataFrame,
# grouped by the unique values in the 'region' column. It gives you a summary of
# the average values for each region in the dataset.
#
insurance_df.groupby('region').mean()
```

```python
# Example 2
# ---
# We can get the first entry for each gender by doing the following
# ---
# NB: You can also get the last one too using last()
# ---
#
insurance_df.groupby('sex').first()
```

```python
# Example 3
# ---
# We can group a single column by another column then apply an aggregate function as shown
# ---
# We group the 'expenses' column in the DataFrame 'insurance_df' by the 'sex' column
# and then calculate the sum of expenses for each group. This operation provides the
# total expenses for each unique value in the 'sex' column, effectively giving the
# total expenses for each gender category in the dataset.
#
insurance_df['expenses'].groupby([insurance_df['sex']]).sum()

# Method 2
# ---
# NB: The groupby output will have an index or multi-index on rows corresponding
# to our chosen grouping variables. To avoid setting this index,
# we pass as_index = False to the groupby operation
# ---
# insurance_df.groupby('sex', as_index = False)['expenses'].sum()
```

```python
# Example 4
# ---
# We can group the region column by another column i.e. bmi then apply an aggregate function
# as shown
# ---
# We group the 'bmi' column in the DataFrame 'insurance_df' by the 'region' column
# and then calculate the mean (average) BMI for each group. This operation provides
# the average BMI for each unique value in the 'region' column, giving insights
# into the average BMI by region in the dataset.
#
insurance_df['bmi'].groupby([insurance_df['region']]).mean()

# Method 2
# ---
# insurance_df.groupby('region', as_index = False)['bmi'].mean()
```

```python
# Example 5
# ---
# We can create a summary statistical table as shown below
# ---
# We group the 'expenses' column in the DataFrame 'insurance_df' by the 'region' column
# and then calculate various descriptive statistics for each group using the describe() function.
# This will provide statistics such as count, mean, standard deviation, minimum, 25th percentile,
# median, 75th percentile, and maximum for the 'expenses' column for each unique value in the 'region' column.
# It gives us a detailed summary of expenses distribution by region in the dataset.
#
insurance_df['expenses'].groupby([insurance_df['region']]).describe()

# Method 2
# ---
# insurance_df.groupby('region', as_index = False)['expenses'].describe()
```

```python
# Example 6
# ---
# Further seeing how we can group a column by another column as shown
# ---
# We group the 'children' column in the DataFrame 'insurance_df' by the 'region' column
# and then calculate the sum of children for each group.
# This operation provides the total number of children for each unique value
# in the 'region' column, giving you the sum of children by region in the dataset.
#
insurance_df['children'].groupby([insurance_df['region']]).sum()

# Method 2
# ---
#
# insurance_df.groupby('region', as_index = False)['children'].sum()
```

```python
# Example 7
# ---
# Applying multiple aggregate functions after grouping a column by another column
# ---
# We group the 'expenses' column in the DataFrame 'insurance_df' by the 'region' column
# and then aggregate multiple statistics using the agg() function.
# In this case, we are calculating the minimum, maximum, and sum of expenses
# for each group based on the unique values in the 'region' column.
# This provides a summary of these three statistics for expenses in each region in the dataset.
#
insurance_df['expenses'].groupby([insurance_df['region']]).agg([min, max, sum])

# Method 2
# insurance_df.groupby(['region'], as_index = False).agg({'expenses': [min, max, sum] })
```

```python
# Example 8
# ---
# We can also apply multiple aggregate functions after grouping a column by another column
# ---
# We group the DataFrame 'insurance_df' by two columns, 'region' and 'sex,'
# while setting as_index to False to keep these columns as regular columns rather than index.
# Then, we aggregate statistics for the 'expenses' column within each group.
# Specifically, we calculate the minimum, maximum, and sum of expenses for each combination
# of 'region' and 'sex,' providing a detailed summary of these statistics for expenses based
# on both region and gender in the dataset.
#
insurance_df.groupby(['region', 'sex'], as_index = False).agg({'expenses': [min, max, sum] })
```

```python
# Example 9
# ---
# Grouping by of multiple columns and then applying an aggregate function
# ---
#
insurance_df['expenses'].groupby([insurance_df['region'], insurance_df['sex']]).sum()

# Method 2
# ---
# insurance_df.groupby(['region', 'sex'], as_index = False)['expenses'].sum()

# Method 3
# ---
# insurance_df.groupby(['region', 'sex']).agg({'expenses': ['sum']})
```
